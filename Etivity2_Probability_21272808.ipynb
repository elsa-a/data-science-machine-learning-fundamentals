```python
# Your code
import random
import numpy as np
import matplotlib.pyplot as plt
```

## Experiment 1 - Uniform Distribution

The following code simulates samples from a rolling dice experiment.
Variables are discrete. 

```python
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

sample_size =10000

dices = list(range(sample_size)) #create an empty list

for i in range(sample_size): #fill the list with randomly generated integers with values ranging 1 to 6.
  dices[i] = random.randint(1,6) 
```

### Histogram for the uniform distribution
The following code uses the data generated in the previous cell and plots it into a histogram showing the frequency of each possible result.

```python
n_bins = 6
# Generate an uniform distribution with data in 'dices'
plt.xlabel('Dice result')
plt.ylabel('Frecuency')
plt.title('Histogram of a Rolling Dice Experiment (Uniform Distribution)')
plt.hist(dices, bins=n_bins, align='mid', color='powderblue') # Plot is centered in the middle, with 'n_bins'
```

### Results for Experiment 1

```python
mean = np.mean(dices) #mean 
std = np.std(dices) #standard deviation 

print("NORMAL DISTRIBUTION\n\nMean:", mean)

print("\nStandard Deviation:", std, "\n")
```

## Experiment 2: Normal distribution

The following code simulates samples from scores achieved by students in a given module. Average grading results are out of 100 and the variables in this case are continous.

```python
normal_data = np.random.normal(loc=75, scale=6.2, size=2000) #loc = mean, scale = std, size = # samples
normal_data[0:20] #print 20 first generated values
```

```python
plt.xlabel('Grades (out of 100)')
plt.ylabel('Frequency')
plt.title('Histogram of Average Grades obtained by students (Normal Distribution)')
plt.hist(normal_data, bins=10, align='mid', rwidth=1, color='powderblue')
```

### Results for Experiment 2

```python
mean = np.mean(normal_data) #mean
std = np.std(normal_data) #standard deviation
print("NORMAL DISTRIBUTION\n\nMean:", mean)
print("\nStandard Deviation:", std, "\n")
```
